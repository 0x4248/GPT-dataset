```python
import csv
from matplotlib import pyplot as plt
```

```python
with open('../dataset/1221/data.csv') as f:
    reader = csv.reader(f)
    next(reader)
    data = list(reader)
```

```python
#plot data
for row in data:
    plt.scatter(row[0], row[1])
```

```python
import random
random.seed(0)
random_points = random.sample(data, 10)
for row in random_points:
    plt.scatter(row[0], row[1], color='red')
```
